```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Define the number of clusters in each dimension
num_clusters_x = 10
num_clusters_y = 10

# Define the size of the clusters
cluster_size = 1

# Create a figure and axis
fig, ax = plt.subplots()

# Loop over each position in the grid
for i in range(num_clusters_x):
    for j in range(num_clusters_y):
        # Compute the position of the cluster
        pos_x = i * cluster_size
        pos_y = j * cluster_size

        # Create a hexagon patch at this position
        hexagon = patches.RegularPolygon((pos_x, pos_y), numVertices=6, radius=0.5*cluster_size,
                                          orientation=np.pi/2, edgecolor='k')

        # Fill the hexagon with a color based on the type of cluster
        if (i + j) % 3 == 0:
            hexagon.set_facecolor('r')
        elif (i + j) % 3 == 1:
            hexagon.set_facecolor('g')
        else:
            hexagon.set_facecolor('b')

        # Add the patch to the axis
        ax.add_patch(hexagon)

# Set the limits of the plot
ax.set_xlim(0, num_clusters_x * cluster_size)
ax.set_ylim(0, num_clusters_y * cluster_size)

# Show the plot
plt.show()
```
